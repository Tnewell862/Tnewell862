```python
# Merge the 2019 and 2020 Guardian results to see which footballers had the largest improvement year on year (Zlatan!)

footballers_guardian_2019 = footballers[(footballers.Publication == "Guardian") & (footballers.Year == 2019)]
footballers_guardian_2020 = footballers[(footballers.Publication == "Guardian") & (footballers.Year == 2020)]
```

```python
# Only interested in Player, Club and Rank columns
footballers_guardian_2019= footballers_guardian_2019[["Player", "Rank"]]
footballers_guardian_2020 = footballers_guardian_2020[["Player", "Rank"]]
```

```python
# merge on player, adding 2019 and 2020 suffixes
merged_years = pd.merge(footballers_guardian_2019, footballers_guardian_2020, on='Player', suffixes=(2019, 2020))
```

```python
# Add new column 'Rank Difference' calculated as the difference in yearly rankings
merged_years["Rank_Difference"] = merged_years["Rank2019"] - merged_years["Rank2020"]

# Finally sort in descending order to see who had a biggest improvement
merged_years.sort_values("Rank_Difference", ascending=False)
```

```python
# Function to calculate cumulative score of footballer based on ranks
# For each year a footballer scores (101 - Rank), inverting the ranks to create a cumulative score.
# If no rank available, footballer scores zero for that year.
# Use numpy to sum the scores

import numpy as np

def calculate_footballer_score(ranks):
    consistency_score = []
    
    for rank in ranks:
        if rank != 0:
            consistency_score.append(101 - rank)
    
    return np.sum(consistency_score)
```

```python
# Use consistency function to aggregate the Rankings, using .groupby and passing function name into .agg
footballers_guardian_rankings = footballers[footballers.Publication == "Guardian"][["Player", "Rank"]]
consistency_table = footballers_guardian_rankings.groupby("Player").agg(calculate_footballer_score)

# Rename Rank columns to Consistency Score using a columns dictionary to map old name to new name
consistency_table = consistency_table.rename(columns={"Rank": "Consistency"})

# Sort values and show the top 10 consistent footballers
consistency_table = consistency_table.sort_values("Consistency", ascending=False)
top10_consistency_table = consistency_table.head(10)
top10_consistency_table
```

```python
# function to plot consistency levels of a given footballer over time using matplotlib
import matplotlib.pyplot as plt

def plot_footballer_consistency(frame, player):
    years = sorted(frame.Year.unique())
    player_rankings = frame[frame.Player == player]

    yearly_ranks = []
    for year in years:
        yearly_ranks_df = player_rankings[player_rankings.Year == year]
        if yearly_ranks_df.empty:
            yearly_ranks.append(0)
        else:
            yearly_ranks.append(yearly_ranks_df.squeeze().Rank)

    # convert yearly rankings to yearly scores
    yearly_scores = []
    for rank in yearly_ranks:
        yearly_scores.append(calculate_footballer_score([rank]))

    yearly_scores

    plt.bar(years, yearly_scores)
    plt.show()
```

```python
# Run function for Harry Kane
plot_footballer_consistency(footballers_guardian, "Harry Kane")
```
